```python
commentaries = []
behind_scenes = []
trailers = []
deleted_scenes = []
for e in film.special_features:
    if 'Deleted Scenes' in e:
        deleted_scenes.append('Y')
    else: 
        deleted_scenes.append('N')
    if 'Behind the Scenes' in e:
        behind_scenes.append('Y')
    else:
        behind_scenes.append('N')
    if 'Trailers' in e:
        trailers.append('Y')
    else:
        trailers.append('N')
    if 'Commentaries' in e:
        commentaries.append('Y')
    else:
        commentaries.append('N')
```

```python
film['deleted scenes'] = deleted_scenes
```

```python
film['behind the scenes'] = behind_scenes
```

```python
film['commentaries'] = commentaries
```

```python
film['trailers'] = trailers
```
